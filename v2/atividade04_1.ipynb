```python
import numpy as np
import matplotlib.pyplot as plt
```

## Exercício 01

### Utilizando método dos quadrados mínimos
- `m = (NΣxy - ΣxΣy) / (NΣx^2 - (Σx)^2)`
- `b = (Σy - mΣx) / N`

```python
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([13, 15, 20, 14, 15, 13, 10])

N = len(x)
m = (N * np.sum(x * y) - np.sum(x) * np.sum(y)) / (N * np.sum(x**2) - (np.sum(x))**2)
b = (np.sum(y) - m * np.sum(x)) / N

def linear_regression(x):
    return m * x + b

plt.scatter(x, y, label='Pontos')
plt.plot(x, linear_regression(x), color='black', label='Reta de ajuste')
plt.title('Ajuste linear - quadrados mínimos')
plt.xlabel('Horas do dia)')
plt.ylabel('Temperatura em °C')
plt.legend()
plt.show()
```

## Exercício 02

```python
x = np.array([0.0, 1.5, 2.5, 3.5, 4.5])
y = np.array([2.0, 3.6, 5.4, 8.1, 12.0])
logy = np.log(y)
coefficients = np.polyfit(x, logy, 1)
a = np.exp(coefficients[1])
b = coefficients[0]

def exponential_fit(x):
    return a * np.exp(b * x)

plt.scatter(x, y, label='Data points')
plt.plot(x, exponential_fit(x), color='red', label='Exponential fit')
plt.title('Least-Squares Exponential Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
print(f'a: {a}, b: {b}')
```

## Exercício 03

```python

# Given data
years = np.array([1950, 1955, 1960, 1965, 1966, 1957, 1958])
accidents = np.array([8200, 9800, 10300, 13100, 13500, 13600, 14500])

# Fit a linear model
linear_fit = np.polyfit(years, accidents, 1)
linear_equation = np.poly1d(linear_fit)

# Predict the number of accidents in 1990
predicted_accidents_1990 = linear_equation(1990)

# Create the plot
plt.scatter(years, accidents, label='Actual Data')
plt.plot(years, linear_equation(years), color='red', label='Linear Fit')

# Plot settings
plt.title('Linear Least-Squares Fit')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.legend()

# Show the plot
plt.show()

# Print the predicted number of accidents in 1990
print(f'Predicted number of accidents in 1990: {predicted_accidents_1990:.2f}')
```

```python
# Given data for part b
accidents_per_10000_vehicles = np.array([1687, 1576, 1396, 1438, 1418, 1384, 1414])

# Fit a quadratic model
quadratic_fit = np.polyfit(years, accidents_per_10000_vehicles, 2)
quadratic_equation = np.poly1d(quadratic_fit)

# Predict the number of accidents per 10,000 vehicles in 1990
predicted_accidents_per_10000_vehicles_1990 = quadratic_equation(1990)
print(f'Número de predições de 10000 veículos em 1990: {predicted_accidents_per_10000_vehicles_1990:.2f}')
```

## Exercício 04

```python

t = np.array([0, 1, 2, 3, 5])
p = np.array([1.0, 1.8, 3.3, 6.0, 11.0])
ln_p = np.log(p)
linear_fit = np.polyfit(t, ln_p, 1)
linear_equation = np.poly1d(linear_fit)
a = np.exp(linear_equation(0))
b = linear_fit[0]

def exponential_fit(t):
    return a * np.exp(b * t)

plt.scatter(t, p, label='Data points')
plt.plot(t, exponential_fit(t), color='red', label='fit')
plt.title('Método dos quadrados mínimos')
plt.xlabel('horas')
plt.ylabel('população')
plt.legend()
plt.show()

predicted_population_5_hours = exponential_fit(5)
print(f'População em 5 horas: {predicted_population_5_hours:.2f} (em milhares)')
```

## Exercício 05

```python
# Given data
years = np.array([1, 2, 3])
profits = np.array([4, 2, 1])
N = len(years)
m = (N * np.sum(years * profits) - np.sum(years) * np.sum(profits)) / (N * np.sum(years**2) - (np.sum(years))**2)
b = (np.sum(profits) - m * np.sum(years)) / N

def linear_function(x):
    return m * x + b

predicted_profits = linear_function(years)

print("Previsão de ganhos:", predicted_profits)
```

## Exercício 06

```python
years = np.array([1960, 1970, 1980, 1990, 2000])
population = np.array([12600, 14000, 16100, 19100, 23200])
A = np.vstack([np.ones_like(years), years, years**2]).T
b = population
coefficients = np.linalg.solve(np.dot(A.T, A), np.dot(A.T, b))
c, b, a = coefficients

def quadratic_function(t):
    return a * t**2 + b * t + c

predicted_population_2010 = quadratic_function(2010)

print(f"Coeficientes: a={a}, b={b}, c={c}")
print(f"População prevista: {predicted_population_2010:.2f}")
```
